```python

#univariate analysis

plt.figure(figsize = (10, 6))
sns.histplot(df['AGE'].dropna(), kde = True, color = 'skyblue')
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize = (10, 6))
sns.violinplot(x = 'AGE', y='SMOKING', data=df)
plt.show()
```

```python
df['LUNG_CANCER'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'], explode=(0, 0.1), startangle=90)
```
